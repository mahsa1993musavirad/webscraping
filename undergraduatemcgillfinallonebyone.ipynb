```python

from selenium import webdriver
import pandas as pd
driver = webdriver.Chrome('Desktop/chromedriver.exe')
driver.get('https://www.mcgill.ca/undergraduate-admissions/programs?f%5B0%5D=field_areas_interest%3A18')
elements = driver.find_elements_by_css_selector('.views-field-title a')
links = []
titles = []
for i in range(len(elements)):
    links.append(elements[i].get_attribute('href'))
    titles.append(elements[i].text)

#block-system-main > div > div > div > div.item-list > ul
pages = driver.find_elements_by_css_selector('#block-system-main li')
print(pages)
print(len(pages))
for limit in range((len(pages))):
    try:
        
        next = driver.find_element_by_css_selector('.pager-next a')
        next.click()
        elements = driver.find_elements_by_css_selector('.views-field-title a')
        for i in range(len(elements)):
            links.append(elements[i].get_attribute('href'))
            titles.append(elements[i].text)
    except:
        pass
print(links)
print(titles)
print(len(links))
print(len(titles))
```

```python
data = []
overview = []
basic_inf_quebec = []
basic_inf_ontorio = []
basic_inf_other_canadian = []
basic_inf_us = []
basic_inf_other = []
for link in links:
    driver.get(link)
    tables = pd.read_html(link)
    basic_inf_quebec.append(tables[0])
    basic_inf_ontorio.append(tables[1])
    basic_inf_other_canadian.append(tables[2])
    basic_inf_us.append(tables[3])
    basic_inf_other.append(tables[4])
    ov = driver.find_element_by_css_selector('.even:nth-child(1) .even .even , .even:nth-child(1) .even .even p')
    overview.append(ov.text)
    print(overview)
data.append([titles, links, basic_inf_quebec, basic_inf_ontorio, basic_inf_other_canadian, basic_inf_us, basic_inf_other, overview])
```

```python
serielink = pd.Series(links)
serietitle = pd.Series(titles)
seriebasic_inf_quebec = pd.Series(basic_inf_quebec)
seriebasic_inf_ontorio = pd.Series(basic_inf_ontorio)
seriebasic_inf_other_canadian = pd.Series(basic_inf_other_canadian)
seriebasic_inf_us = pd.Series(basic_inf_us)
seriebasic_inf_other = pd.Series(basic_inf_other)
serieoverview = pd.Series(overview)

frame = {'titles':serietitle
         , 'links': serielink,
         'basic_inf_quebec': seriebasic_inf_quebec
         , 'basic_inf_ontorio': seriebasic_inf_ontorio
         , 'basic_inf_other_canadian': seriebasic_inf_other_canadian, 
         'basic_inf_us': seriebasic_inf_us,
         'basic_inf_other':seriebasic_inf_other,
         'overview':serieoverview}
```
